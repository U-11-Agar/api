```python
import requests
import time
import pandas as pd

# Alpha Vantage API key
API_KEY = 'YOUR_API_KEY'

# Base URL for Alpha Vantage API
BASE_URL = 'https://www.alphavantage.co/query?'

# Function to fetch forex exchange details
def get_forex_exchange_details(symbol):
    """
    Fetches live stock details of forex exchange for a given symbol.

    Args:
        symbol (str): Symbol of the forex exchange (e.g. 'EURUSD', 'GBPJPY', etc.)

    Returns:
        dict: A dictionary containing the live stock details of the forex exchange.
    """
    # API parameters
    params = {
        'function': 'CURRENCY_EXCHANGE_RATE',
        'from_currency': symbol[:3],
        'to_currency': symbol[3:],
        'apikey': API_KEY
    }

    try:
        # Make API request
        response = requests.get(BASE_URL, params=params)
        data = response.json()

        # Extract relevant data
        forex_exchange_details = {
            'symbol': symbol,
            'exchange_rate': float(data['Realtime Currency Exchange Rate']['5. Exchange Rate']),
            'timestamp': data['Realtime Currency Exchange Rate']['6. Last Refreshed']
        }

        return forex_exchange_details
    except Exception as e:
        print(f'Error fetching forex exchange details: {e}')
        return None

# Example usage
symbol = 'EURUSD'  # Example forex exchange symbol

# Create an empty DataFrame
df = pd.DataFrame(columns=['Symbol', 'Exchange Rate', 'Timestamp'])

while True:
    forex_exchange_details = get_forex_exchange_details(symbol)
    if forex_exchange_details:
        # Append fetched data to DataFrame
        df = df.append(forex_exchange_details, ignore_index=True)
        print(df)
    time.sleep(15)  # Sleep for 15 seconds
```
